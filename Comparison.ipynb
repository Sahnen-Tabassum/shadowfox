```python
import pandas as pd
import numpy as np
#Create sample dataset
data = {
    'Year': [2018, 2019, 2020, 2021, 2022, 2023],
    'Patients': [120, 150, 170, 160, 180, 200],
    'SurvivalRate': [60, 62, 65, 63, 68, 70],
    'CancerType': ['Lung', 'Breast', 'Colon', 'Liver', 'Leukemia', 'Skin']
}

df = pd.DataFrame(data)
```

```python
#LINE PLOT
df.plot(
    x='Year',
    y='Patients',
    kind='line',
    title="Patients Over Years",
    color='green',
    marker='o'
)
```

```python
#BAR CHART
df.plot(
    x='CancerType',
    y='Patients',
    kind='bar',
    title="Patients by cancer Type",
    color=['red','blue','green','purple','orange','cyan']
)
```

```python
#PIE CHART
df.set_index('CancerType')['Patients'].plot(
    kind='pie',
    autopct='%1.1f%%',
    title="Patient Distribution by Cancer Type",
    colors=['gold','lightblue','lightgreen','pink','violet','coral']
)
```

```python
#SCATTER PLOT
df.plot(
    x='Patients',
    y='SurvivalRate',
    kind='scatter',
    title="Patients vs Survival Rate",
    color='purple'
)
```

```python
#HISTOGRAM
df['Patients'].plot(
    kind='hist',
    bins=5,
    title="Distribution of Patients",
    color='lightpink'
)
```

```python
#BOX PLOT
df[['Patients','SurvivalRate']].plot(
    kind='box',
    title="Box Plot of Data",
    color=dict(boxes='blue',whiskers='black',medians='red',caps='green')
)
```
